### Problem 1 
### There are various stocks for which we have collected a data set, which all stocks are
### apparently similar in performance

```python
data['NEW_DATE'] = pd.to_datetime(data['DATE'],unit='s')
```

```python
data.drop(columns=['DATE','SP500'],axis=1,inplace=True)
```

```python
data.head() #Returns the first 5 rows of stocks dataframe
```

```python
data.set_index('NEW_DATE',inplace=True)
```

```python
data.head()
```

```python
data_T = data.transpose() #Creating transpose of the df dataframe
```

```python
data_T.head() #Returns the first 5 rows of df_transpose dataframe
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

pca_3 = PCA(n_components=3)
```

```python
pca_data = pca_3.fit_transform(data_T)
```

```python
no_of_clusters = []
inertia = []
for i in range(2,10):
    kmeans_clf=KMeans(n_clusters=i)
    kmeans_clf.fit(pca_data)
    no_of_clusters.append(i)
    inertia.append(kmeans_clf.inertia_)
```

```python
#Plot to find number of clusters (Elbow Method)
plt.plot(no_of_clusters,inertia)
plt.show()
```

```python
#Initializing and fitting KMeans
kmeans_clf = KMeans(n_clusters = 6)
kmeans_clf.fit(pca_data)
```

```python
#Predicted values using KMeans
y_pred = kmeans_clf.predict(pca_data)
```

```python
#Scatter Plot
plt.scatter(pca_data[:,0],pca_data[:,1],c=y_pred)
plt.show()
```

```python
data_T['y_pred'] = y_pred
```

```python
# Below are various stocks which are similar in performance
data_T['y_pred'].value_counts()
```

```python
# Stocks which belongs to Cluster 1 
data_T[data_T['y_pred']==1]
```
